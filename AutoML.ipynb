```python
# Import the necessary libraries
import h2o
from h2o.automl import H2OAutoML
```

```python
# Initialize H2O
h2o.init()
```

```python
# Load the dataset
dataset = h2o.import_file("iris.csv")
```

```python
dataset.show()
```

```python
# Specify the target column
target_col = "species"

# Split the data into training and testing sets
train, test = dataset.split_frame(ratios=[0.7], seed=42)

# Define the predictor columns
predictors = dataset.columns[:-1]
```

```python
# Initialize and train the AutoML model
aml = H2OAutoML(max_models=10, seed=42)
aml.train(x=predictors, y=target_col, training_frame=train)
```

```python
# Get the best model from AutoML
best_model = aml.leader

# Evaluate the best model on the test data
perf = best_model.model_performance(test_data=test)
print("Best Model Performance:")
print(perf)
```

```python
# Compare all models' performances
leaderboard = aml.leaderboard
print("Leaderboard of All Models:")
print(leaderboard)
```

```python
# Shutdown H2O
h2o.cluster().shutdown()
```
